```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('../Database/kp_classify_blur.csv')
df2 = pd.read_csv('../Database/kp_classify_filtered.csv')
```

```python
match_rate1 = df1['match'].mean()
match_rate2 = df2['match'].mean()
```

```python
print(f'Match rate for Dataset 1: {match_rate1:.2%}')
print(f'Match rate for Dataset 2: {match_rate2:.2%}')
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df1[df1['match'] == 1], x='blur', label='Raw Data', fill=True)
sns.kdeplot(data=df2[df2['match'] == 1], x='blur', label='Model Filtered Data', fill=True)
plt.title('KDE Plot of Blur for Match = 1')
plt.xlabel('Blur')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df1[df1['match'] == 1], x='size', label='Raw Data', fill=True)
sns.kdeplot(data=df2[df2['match'] == 1], x='size', label='Model Filtered Data', fill=True)
plt.title('KDE Plot of Blur for Match = 1')
plt.xlabel('Size')
plt.ylabel('Density')
plt.legend()
plt.show()
```
